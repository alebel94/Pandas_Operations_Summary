# Data Types and Missing Values

```python
import pandas as pd
df_path = 'data_files/winemag-data-130k-v2.csv'
# speccifies a data path
wine_df = pd.read_csv(df_path, index_col = 0)
# makes a dataframe from that path. May need to use index_col = 0 to drop first column
wine_df
```

### DTypes

#### The following are the pandas column data types

```python
all_rows = [['object', 'str or mixed', 'string_, unicode_, mixed types', 'Text or mixed numeric and non-numeric values'],
['int64', 'int', 'int_, int8, int16, int32, int64, uint8, uint16, uint32, uint64', 'Integer numbers'], 
['float64', 'float', 'float_, float16, float32, float64', 'Floating point numbers'], 
['bool', 'bool', 'bool_', 'True/False values'], 
['datetime64', 'datetime', 'datetime64[ns]', 'Date and time values'], 
['timedelta[ns]', 'NA', 'NA', 'Differences between two datetimes'], 
['category', 'NA', 'NA', 'Finite list of text values']]

data_types = pd.DataFrame(all_rows, columns=['Pandas_dtype', 'Python_type', 'NumPy_type', 'Usage'])
data_types.style.hide_index()
```

```python
wine_df.points.dtype
# data type for a column in a DataFrame or a Series is known as the dtype
# Different data types can be seen above
```

```python
wine_df.dtypes
# Can also get all columns without specifying the column
```

```python
wine_df2 = wine_df.points.astype('float64')
wine_df2
# One way to change a column's datatype. It will return a column
# wine_df2 has to be printed to see the change. wine_df remains as int64
# can test doing wine_df2.dtypes
```

```python
wine_df2 = wine_df.astype({'points':'float64'})
wine_df2
# This way, the column has changed data type and returns a dataframe not a column
# wine_df2 has to be printed to see the change. wine_df remains as int64
# can test doing wine_df2.dtypes
```

```python
wine_df.index.dtype
# Index has its own data type, int64
```

### Missing data

#### Missing values are given a value NaN (No a Number). NaN is always float64

```python
wine_df[pd.isnull(wine_df.country)]
# Selects all rows that contain a NaN in the 'country" column
```

```python
wine_df[pd.notnull(wine_df.country)]
# Selects all rows that contain a value, not NaN, in the 'country" column
```

```python
wine_df.region_2.fillna("Unknown")
# replaces NaN values in the region_2 for dtype object "unknown"
# another strategy is to fill NaN with the first non-null value that appears sometime after 
    # the given record in the database. That is known as backfill strategy
```

```python
wine_df.taster_twitter_handle.replace("@kerinokeefe", "@kerino")
# .replace() replaces the first value with the second value in the taster_twitter_handle column
# Useful when replacing missing data
```

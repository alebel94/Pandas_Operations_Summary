### Data selection (Indexing, Selecting & Assigning)

```python
#wine_df = df.drop(axis=0)
#drops missing value rows
```

```python
wine_df.country
# accesses the property(country column) of an object(df) by accessing it as an attribute
#Attributes are like methods, but instead of transforming the variables or data they are...
    # ...used to give you more information about the data
# In other words, it gives info about the Series (row) by selecting it 
```

```python
wine_df['country']
# Another way to access Series (row) of a dataframe
# This can select Series that have spaces for names
```

```python
wine_df['country'][0]
# A way to index select a cell 
```

 ##### Pandas indexing has two padigrams:
 ##### iloc: index-based selection: selecting data based on its numerical position in the data
 ##### loc: label-based selection: selecting data based on index value, not its position

```python
wine_df.iloc[0]
# iloc can be used to select rows, columns, or cells by index position.
# First index is always rows, second is always colunns
# Here we are selecting the first row. No need to specify column
```

```python
wine_df.iloc[:, 0]
# This is an example selecting the first column instead of row
```

```python
wine_df.iloc[1:3, 0]
# just like a python list iloc can select a range, last value EXCLUDED
```

```python
wine_df.iloc[[0, 4, 7], 0]
# iloc can also select a list of rows or columns using []
```

```python
wine_df.iloc[-5:]
# iloc can also select going back. the -5 is inclusive
```

```python
wine_df.loc[1:3, 'country']
# loc selects by value not location
# just like iloc, a range can be selected. However, this time the last 
    # value is INCLUDED
```

```python
wine_df.loc[1, 'points':'price']
# It is easier to understand the reason it is INCLUDED with words as values.
# If the last value was excluded then the range would have to be writen:
# wine_df.loc[1, 'points':'pricf']
# This is because f comes after e in the alphabet so to reach e, f would have to be excluded
```

```python
wine_df.set_index("title")
# A column in the dataframe can be moved and set as the index 
```

```python
wine_df.country == 'Italy'
# conditional selection can be used (==) to return a Series of booleans that match
```

```python
wine_df.loc[wine_df.country == 'Italy']
# loc can be used with a conditional selection to find rows that contain a value
# unlike before, when using loc you get a dataframe not Series
# The original dataframe contains 129,971 rows, this one contains 19,540... only italy entries
```

```python
wine_df.loc[(wine_df.country == 'Italy') & (wine_df.points >= 90)]
# the 'and' logic operator (&) can be used to filter further
# this more filtered dataframe contains only 6,648 rows
```

```python
wine_df.loc[(wine_df.country == 'Italy') | (wine_df.points >= 90)]
# just like &, the 'or' logic operator (|) can be used to filter further as well
# this more filtered dataframe contains 61,937 rows, more than the 19,540 for just the
    # italy conditional selection because it includes all italy rows and points >= 90
```

```python
wine_df.loc[wine_df.country.isin(['Italy', 'France'])]
# .isin([]) is one of two built in pandas conditional selectors
# this selects values from a list and displays the rows with those values
```

```python
wine_df.loc[wine_df.price.notnull()]
# .notnull() is the other built in pandas conditional selector (companion notnull also exists)
# this selects rows that include a value for that specific column ('price' in this case)
# 120,975 of 129,971 rows contain a value in the 'price' column
```

```python
wine_df.loc[wine_df.price.isnull()]
# .isnull() is the companion of notnull and selects the opposite
# this selects rows that include a null (NaN) for that specific column ('price' in this case)
# 8,996 of 129,971 rows contain a null (NaN) not value in the 'price' column
```

```python
wine_df['critic'] = 'everyone'
wine_df['critic']
# Assigning one constant value for all values of a new column 'critic'
# This overwrites the dataframe
```

```python
wine_df['index_backwards'] = range(len(wine_df), 0, -1)
wine_df['index_backwards']
# another way to assign values to a new column 'index_backwards' can be with an iterable
# This will also overwrite the dataframe
```

### Summary Functions and Maps to reformat data

```python
wine_df.points.describe()
# Gives statistic information... In this case of a specific column 'points' with numeric data
```

```python
wine_df.taster_name.describe()
# With strings this is the data obtain instead
```

```python
wine_df.points.mean()
# We can obtain the mean from a numeric column. Error for string column
```

```python
wine_df.taster_name.unique()
# We can obtain unique values for numeric or string columns
# Individual values 
```

```python
wine_df.taster_name.value_counts()
# We can obtain value counts for numeric or string columns
# Number of occurances per individual values
```

```python
wine_df_points_mean = wine_df.points.mean()
wine_df.points.map(lambda p: p - wine_df_points_mean)
# map() function is the first of two maping functions
# Function passed to map (lambda in this case) should expect a single value from the Series
    # map() returns a new Series transformed by the function passed. Returns a Series
# creates a new dataframe, dows not overwrite wine_df
```

```python
wine_df_points_mean = wine_df.points.mean()
wine_df_price_mean = wine_df.price.mean()

def remean_points(row):
    row.points = row.points - wine_df_points_mean
    row.price = row.price - wine_df_price_mean
    return row

wine_df.apply(remean_points, axis='columns')
# apply() is the second mapping function
# apply works like map() but for every row in the dataframe. Returns a dataframe
# creates a new dataframe, does not overwrite wine_df
```

```python
wine_df_points_mean = wine_df.points.mean()
wine_df.points - wine_df_points_mean
# This operation returns the same values as the map() function seen above
# This operator are faster than map() or apply() because it uses speed ups
    #built into pandas
# however, it is not as flexible as map() or apply(), which can do more advanced
    # things, like applying conditional logic
```

```python
wine_df.country + " - " + wine_df.region_1
# Another example of this operation returns combined data from two columns
# This operator are faster than map() or apply() because it uses speed ups
    #built into pandas
# however, it is not as flexible as map() or apply(), which can do more advanced
    # things, like applying conditional logic
```

### Grouping and Sorting

##### Groupwise Analysis

```python
wine_df.groupby('points').points.count()
# Doing a groupby and count() is the same as doing value_counts()
# Same results using value_counts():
# wine_df.points.value_counts().sort_index()
```

```python
wine_df.groupby('points').price.min()
# groupby() can be used with any of the summary functions
```

```python
wine_df.groupby('winery').apply(lambda df: df.title.iloc[0])
# groupme() is a slice of the dataframe
# This DataFrame is accessible using the apply() method, and then it can be manipulated any way we see fit. 
# example shows one way of selecting the name of the first wine reviewed from each winery in the dataset:
```

```python
wine_df.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
# groupby() can work for more than one column
# idxmax() finds the index of the max value along the index axis (returns the row index of max value)
```

```python
wine_df.groupby(['country']).price.agg([len, min, max])
# agg() can be used with groupby() to run a bunch of different functions, in this case len, min, max
```

##### Multi-Indexes

```python
countries_reviewed = wine_df.groupby(['country', 'province']).description.agg([len])
countries_reviewed
# 
```

```python
mi = countries_reviewed.index
print(mi, '\n \n \n', type(mi))
```

```python
countries_reviewed.reset_index()
```
